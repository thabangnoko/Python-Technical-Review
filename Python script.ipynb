```python
import pandas as pd
```

```python
dataset = pd.read_csv('Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv')
dataset
```

```python
dataset.columns
```

```python
dataset_subset = dataset[['YearStart', 'LocationAbbr', 'LocationDesc', 'Datasource', 'Class', 'Question',
       'Data_Value', 'Low_Confidence_Limit', 'High_Confidence_Limit ',
       'Sample_Size', 'ClassID',  'QuestionID', 'LocationID', 'StratificationCategory1',
       'Stratification1', 'StratificationCategoryId1', 'StratificationID1']]
```

```python
pd.isna(dataset_subset)
```

```python
clean_dataset = dataset_subset.dropna()
```

```python
#Seperating the data into 3 dataframes for the categories Physical activity, Obesity/weight status, Fruits and Vgetables
PA = clean_dataset[clean_dataset['ClassID'].str.contains('PA')]
OWS = clean_dataset[clean_dataset['ClassID'].str.contains('OWS')]
FV = clean_dataset[clean_dataset['ClassID'].str.contains('FV')]
```

```python
PA_150_or_75 = PA[PA['QuestionID'].str.contains('Q043')]
PA_150_or_75_2 = PA[PA['QuestionID'].str.contains('Q044')]
PA_150_2 = PA[PA['QuestionID'].str.contains('Q045')]
PA_ms_2 = PA[PA['QuestionID'].str.contains('Q046')]
PA_na = PA[PA['QuestionID'].str.contains('Q047')]

PA_150_or_75_age = PA_150_or_75[PA_150_or_75['StratificationCategoryId1'].str.contains('AGEYR')]
PA_150_or_75_2_age = PA_150_or_75_2[PA_150_or_75_2['StratificationCategoryId1'].str.contains('AGEYR')]
PA_150_2_age = PA_150_2[PA_150_2['StratificationCategoryId1'].str.contains('AGEYR')]
PA_ms_2_age = PA_ms_2[PA_ms_2['StratificationCategoryId1'].str.contains('AGEYR')]
PA_na_age = PA_na[PA_na['StratificationCategoryId1'].str.contains('AGEYR')]

PA_150_or_75_yearly = PA_150_or_75.groupby('YearStart')[['Data_Value']].mean()
PA_150_or_75_2_yearly = PA_150_or_75_2.groupby('YearStart',)[['Data_Value']].mean()
PA_150_2_yearly = PA_150_2.groupby('YearStart')[['Data_Value']].mean()
PA_ms_2_yearly = PA_ms_2.groupby('YearStart')[['Data_Value']].mean()
PA_na_yearly = PA_na.groupby('YearStart')[['Data_Value']].mean()

PA_150_or_75_age2 = PA_150_or_75_age.groupby('Stratification1')[['Data_Value']].mean()
PA_150_or_75_2_age2 = PA_150_or_75_2_age.groupby('Stratification1')[['Data_Value']].mean()
PA_150_2_age2 = PA_150_2_age.groupby('Stratification1')[['Data_Value']].mean()
PA_ms_2_age2 = PA_ms_2_age.groupby('Stratification1')[['Data_Value']].mean()
PA_na_age2 = PA_na_age.groupby('Stratification1')[['Data_Value']].mean()
```

```python
OWS_Overweight = OWS[OWS['QuestionID'].str.contains('Q036',)]
OWS_Obese = OWS[OWS['QuestionID'].str.contains('Q037')]

OWS_Overweight_year = OWS_Overweight.groupby('YearStart')[['Data_Value']].mean()
OWS_Obese_year = OWS_Obese.groupby('YearStart')[['Data_Value']].mean()

OWS_Overweight_state = OWS_Overweight.groupby(['LocationAbbr', 'YearStart'])[['Data_Value']].mean()

OWS_Overweight_AGE = OWS[OWS['StratificationCategoryId1'].str.contains('AGEYR')]
OWS_Obese_AGE = OWS[OWS['StratificationCategoryId1'].str.contains('AGEYR')]

OWS_Overweight_AGE2 = OWS_Overweight_AGE.groupby('Stratification1')[['Sample_Size']].mean()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output

import dash
from dash import dcc
from dash import html

app = dash.Dash(__name__)

OWS_Obese_state = OWS_Obese.groupby(['LocationAbbr','YearStart'])[['Data_Value']].mean()
OWS_Obese_state.reset_index(inplace=True)
```

```python
app.layout = html.Div([html.H1("Dashboard for obesity rates in the USA", style={'text-align': 'center'}),

    dcc.Dropdown(id="slct_year",
                 options=[
                     {"label": "2011", "value": 2011},
                     {"label": "2012", "value": 2012},
                     {"label": "2013", "value": 2013},
                     {"label": "2014", "value": 2014},
                     {"label": "2015", "value": 2015},
                     {"label": "2016", "value": 2016},
                     {"label": "2017", "value": 2017},
                     {"label": "2018", "value": 2018},
                     {"label": "2019", "value": 2019},
                     {"label": "2020", "value": 2020}],
                 multi=False,
                 value=2011,
                 style={'width': "40%"}
                 ),

    html.Div(id='output_container', children=[]),
    html.Br(),

    dcc.Graph(id='map', figure={})

])
```

```python
@app.callback(
    [Output(component_id='output_container', component_property='children'),
     Output(component_id='map', component_property='figure')],
    [Input(component_id='slct_year', component_property='value')])

def update_graph(option_slctd):
    print(option_slctd)
    print(type(option_slctd))

    container = "The chosen year was: {}".format(option_slctd)

    data = OWS_Obese_state.copy()
    data = data[data["YearStart"] == option_slctd]

    
    fig = px.choropleth(data_frame=data,locationmode='USA-states',locations='LocationAbbr',scope="usa",color='Data_Value',hover_data=['LocationAbbr', 'Data_Value'],color_continuous_scale=px.colors.sequential.YlOrRd,labels={'Data_Value': '%'},template='plotly_dark')
    
    
    fig = go.Figure(data=[go.Choropleth(locationmode='USA-states',locations=data['LocationAbbr'], z=data["Data_Value"].astype(float),colorscale='bluered',)])
    
    fig.update_layout(title_text="USA",title_xanchor="center",title_font=dict(size=24),title_x=0.5, geo=dict(scope='usa'),)
    return container, fig
```
